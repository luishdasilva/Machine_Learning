```python
# Crie um modelo de classificação de imagens, a partir da importação da base de dados do
# do Sci-kit Learn.
import matplotlib.pyplot as plt

from sklearn import datasets, svm, metrics
from sklearn.model_selection import train_test_split 


# a) Utilize a base de dados Mnist dataset 

digito = datasets.load_digits()

_, figura = plt.subplots(nrows=1, ncols=4, figsize=(10, 3))
for ax, image, label in zip(figura, digito.images, digito.target):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    ax.set_title('Training: %i' % label)




```

```python
# b) divida a base de treinamento e teste em 80/20
nn = len(digito.images)
comp = digito.images.reshape((nn, -1))

X_train, X_test, y_train, y_test = train_test_split(
    comp, digito.target, test_size=0.2, shuffle=False)
```

```python

# c) Utilize a técnica de cross validation (k-fold=5)
import numpy as np
from sklearn.model_selection import KFold

kf = KFold(n_splits=5)

for X_train, X_test in kf.split(comp):
    print (X_train, X_test)
```

```python
# d)
nn = len(digito.images)
comp = digito.images.reshape((nn, -1))

X_train, X_test, y_train, y_test = train_test_split(
    comp, digito.target, test_size=0.2, shuffle=False)

g = svm.SVC(gamma=0.001)

g.fit(X_train, y_train)

pre = g.predict(X_test)

print(f"Classification report for classifier {kf}:\n"
      f"{metrics.classification_report(y_test, pre)}\n")
```

```python
#e)
disp = metrics.plot_confusion_matrix(g, X_test, y_test)
disp.figure_.suptitle("Confusion Matrix")
print(f"Confusion matrix:\n{disp.confusion_matrix}")

plt.show()
```
